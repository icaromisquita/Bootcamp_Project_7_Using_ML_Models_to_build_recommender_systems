## Create the similarity matrix

In 3 simple steps:

1. Create the big users-items table

2. Replace NaNs with zeros

3. Compute pairwise cosine similarities

### 1. Create the big users-items table.
We are just reshaping (pivoting) the data, so that we have users as rows and movies as columns. We need the data to be in this shape to compute similarities between users in the next step.

```python
import numpy as np
import pandas as pd
movies = pd.read_csv("https://raw.githubusercontent.com/icaromisquita/WBS-Project-8---Reccomenders/main/movies.csv")
tags = pd.read_csv("https://raw.githubusercontent.com/icaromisquita/WBS-Project-8---Reccomenders/main/tags.csv")
links = pd.read_csv("https://raw.githubusercontent.com/icaromisquita/WBS-Project-8---Reccomenders/main/links.csv")
ratings = pd.read_csv("https://raw.githubusercontent.com/icaromisquita/WBS-Project-8---Reccomenders/main/ratings.csv")
```

```python
movies_crosstab = pd.pivot_table(data=ratings, values='rating', index='userId', columns='movieId')
movies_crosstab.head(10)
```

```python
ratings["userId"].count()
```

2. Replace NaNs with zeros

```python
movies_crosstab.fillna(0, inplace=True)
movies_crosstab.head()
```

3. Compute cosine similarities

```python
from sklearn.metrics.pairwise import cosine_similarity

movies_similarities = pd.DataFrame(cosine_similarity(movies_crosstab),
                                 columns=movies_crosstab.index, 
                                 index=movies_crosstab.index)
movies_similarities.head()
```

#Building the recommender step by step:
Let's focus on one random user (userId 1) and compute the recommendations only for this user, as an example. Then, we will build a function that can compute recommendations for any users. We will follow these steps:

Compute the weights.

Find restaurants user userId 1 has not rated.

Compute the ratings user userId 1 would give to those unrated restaurants.

Find the top 5 movies from the rating predictions.

1. Compute the weights
Here we will exclude user userId 1 using .query().

```python
# compute the weights for one user
user_id = 1

weights = (
    movies_similarities.query("userId!=@user_id")[user_id] / sum(movies_similarities.query("userId!=@user_id")[user_id])
          )
weights.head()
```

```python
weights.sum()
```

2. Find movies user 1 has not rated.
We will exclude our user, since we don't want to include them on the weights.

```python
movies_crosstab.loc[user_id,:]==0
```

```python
# select restaurants that the inputed user has not visited
not_watched = movies_crosstab.loc[movies_crosstab.index!=user_id, movies_crosstab.loc[user_id,:]==0]
not_watched.T
```

3. Compute the ratings user 1 would give to those unrated movies.

```python
# dot product between the not-visited-restaurants and the weights
weighted_averages = pd.DataFrame(not_watched.T.dot(weights), columns=["predicted_rating"])
weighted_averages.head()
```

4. Find the top 5 movies from the rating predictions

```python
recommendations = weighted_averages.merge(movies, left_index=True, right_on="movieId")
recommendations.sort_values("predicted_rating", ascending=False).head()
```
